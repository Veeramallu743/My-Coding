###  Model Building

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dense, Flatten
```

```python
feed = Input(shape = (100, 100,3))
```

```python
cv1 = Conv2D(32, kernel_size = (3,3), activation = "relu", strides = (1,1),  padding= "same")(feed)
```

```python
max1 = MaxPooling2D(pool_size=(2,2), padding="valid")(cv1)
```

```python
cv2 = Conv2D(32, kernel_size = (3,3), activation = "relu", strides = (1,1),  padding= "same")(max1)
```

```python
max2 = MaxPooling2D(pool_size=(2,2), padding="valid")(cv2)
```

```python
cv3 = Conv2D(32, kernel_size = (3,3), activation = "relu", strides = (1,1),  padding= "same")(max2)
```

```python
max3 = MaxPooling2D(pool_size=(2,2), padding="valid")(cv3)
```

```python
flat = Flatten()(max3)
```

```python
den1 = Dense(256, activation = "relu")(flat)
```

```python
den2 = Dense(128, activation = "relu")(den1)
```

```python
den3 = Dense(256, activation = "relu")(den2)
```

```python
out = Dense(2, activation = "sigmoid")(den3)
```

```python
fmodel = Model(inputs = feed, outputs = out)
```

```python
fmodel.summary()
```

```python
fmodel.compile( optimizer = "adam",
                loss= "sparse_categorical_crossentropy",
                metrics = ["accuracy"])
```

```python
fmodel.fit(x_train, y_train, epochs = 20)
```

```python
fmodel.evaluate(x_test, y_test)
```
