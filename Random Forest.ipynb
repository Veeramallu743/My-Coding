```python
cols = [' residential_assets_value', ' commercial_assets_value' , ' bank_asset_value']
```

```python
for c in cols:
    print(dfloan[c])
```

```python
for c in cols :
    sk = dfloan[c].skew().round()
    if sk ==0 or sk == -0:
        print(c, " : Normal Distribution")
    else: 
        print(c, ": Skewed Distribution")
```

```python
for c in cols :
    sk = dfloan[c].skew().round()
    if sk ==0 or sk == -0:
        m = dfloan[c].mean()
        s = dfloan[c].std()
        lowerlimit = m - 3*s
        upperlimit = m + 3*s
        dfloan[c] = dfloan[c].map(lambda x : lowerlimit if x < lowerlimit else upperlimit if x > upperlimit else x)
    else: 
        q1 = dfloan[c].quantile(0.25)
        q3 = dfloan[c].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - (1.5 * iqr)
        upper = q3 + (1.5 * iqr)
        dfloan[c] = dfloan[c].map(lambda x : lower if x < lower else upper if x > upper else x) 
```

### Auto ML

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
dt = DecisionTreeClassifier()
sv = SVC()
lr = LogisticRegression()
rfc = RandomForestClassifier()
ada = AdaBoostClassifier()
grad= GradientBoostingClassifier()
```

```python
model = [dt, sv, lr, rfc, ada, grad]
```

```python
model
```

```python
for m in model :
    print(m)
```

```python
sclist = []
for m in model :
    m.fit(x_train, y_train)
    sc = m.score(x_test, y_test)
    print(m, end= "==>")
    print(sc)
    sclist.append(sc)
print("====================================")    
print(sclist)
```
