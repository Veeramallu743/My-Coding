###  Model Building

```python
seq_model = Sequential()
```

### First Convolutional Layer

```python
seq_model.add( Conv2D(64, kernel_size = (3,3), strides = (1,1), padding= "same", 
                      activation = "relu", input_shape = (96, 96, 3))) 
```

```python
seq_model.add( MaxPooling2D(pool_size=(2,2)))
seq_model.add( Dropout (0.25))
```

```python
seq_model.add(BatchNormalization(axis= -1))
```

### Second Convolutional Layer

```python
seq_model.add( Conv2D(128, kernel_size = (3,3), strides = (1,1), padding= "same", activation = "relu")) 
```

```python
seq_model.add( MaxPooling2D(pool_size=(2,2)))
seq_model.add( Dropout (0.25))
```

```python
seq_model.add(BatchNormalization(axis= -1))
```

### Third Convolutional Layer

```python
seq_model.add( Conv2D(128, kernel_size = (3,3), strides = (1,1), padding= "same", activation = "relu")) 
```

```python
seq_model.add( MaxPooling2D(pool_size=(2,2)))
seq_model.add( Dropout (0.25))
```

```python
seq_model.add(BatchNormalization(axis= -1))
```

### Feed forward network

```python
seq_model.add(Flatten())
```

```python
seq_model.add( Dense (256, activation="relu"))
seq_model.add( Dense (512, activation="relu"))
seq_model.add( Dense (256, activation="relu"))
seq_model.add( Dense (2, activation="sigmoid"))
```

```python
seq_model.summary()
```

```python
seq_model.compile( optimizer = "adam",
                loss= "sparse_categorical_crossentropy",
                metrics = ["accuracy"])
```

```python
seq_model.fit(x_train, y_train, epochs = 20)
```

```python
seq_model.evaluate(x_test, y_test)
```
