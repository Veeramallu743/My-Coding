```python
for i in range(61) :
    sk = df[i].skew().round()
    print(i, "==>" , sk)
    if sk ==0 or sk == -0:
        print(i, " : Normal Distribution")
    else: 
        print(i, ": Skewed Distribution")
```

```python
for i in range(61) :
    sk = df[i].skew().round()
    if sk ==0 or sk == -0:
        m = df[i].mean()
        s = df[i].std()
        lowerlimit = m - 3*s
        upperlimit = m + 3*s
        df[i] = df[i].map(lambda x : lowerlimit if x < lowerlimit else upperlimit if x > upperlimit else x)
    else: 
        q1 = df[i].quantile(0.25)
        q3 = df[i].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - (1.5 * iqr)
        upper = q3 + (1.5 * iqr)
        df[i] = df[i].map(lambda x : lower if x < lower else upper if x > upper else x) 
```
